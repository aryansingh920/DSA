```python
import heapq

class TopKPrices:
    def __init__(self, k: int):
        self.heapList = []
        self.size = k 
        
        heapq.heapify(self.heapList)
        
    def checkSize(self) -> bool:
        return len(self.heapList) < self.size

    def addPrice(self, price: int) -> None:
        # print(self.heapList)
        if self.checkSize():
            heapq.heappush(self.heapList,price)
        else:
            if price > self.getKthHighest():
                heapq.heapreplace(self.heapList,price)

    def getKthHighest(self) -> int:
        return (self.heapList[0])
```

```python
functions = ["addPrice", "getKthHighest", "addPrice", "getKthHighest",
             "addPrice", "getKthHighest", "addPrice", "getKthHighest",
             "addPrice", "getKthHighest"]

params = [[5], [], [1], [], [10], [], [7], [], [6], []]

kPricesClasses = TopKPrices(3)

res = []

for fn, arg in zip(functions, params):
    method = getattr(kPricesClasses, fn)  # use the actual object
    output = method(*arg)                 # call with args (or none)
    res.append(output)

[print(i) if i else "" for i in res]
```

```python
from typing import List
def insert(intervals: List[List[int]], newInterval: List[int]) -> List[List[int]]:
    pointer1,pointer2 = 0,1

    while(pointer2 < len(intervals)):
        if(intervals[pointer1][0] >= newInterval[0]):
            intervals.insert(pointer1, newInterval)
            break
        if (intervals[pointer1][0] <= newInterval[0]) and (intervals[pointer2][0] >= newInterval[0]):
            intervals.insert(pointer2, newInterval)
            break
        if(pointer2+1 == len(intervals)):
            if (newInterval[0] > intervals[pointer2][0]):
                intervals.append(newInterval)
                break
        pointer1+=1
        pointer2+=1
        
        
    pointer1,pointer2 = 0,1
    while(pointer2 < len(intervals)):
        print(f"pt1 : {intervals[pointer1]}, {pointer1} | ptr2 : {intervals[pointer2]}, {pointer2}")
        if(intervals[pointer1][1] >= intervals[pointer2][0]):
            print("here")
            intervals[pointer1] = [intervals[pointer1][0],max(intervals[pointer1][1],intervals[pointer2][1])]
            intervals.remove(intervals[pointer2])
            pointer1 -= 1
            pointer2 -= 1
            
            
        print("New intervals : ",intervals)
        
        pointer1+=1
        pointer2+=1
    
    return intervals
        
```
