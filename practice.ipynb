```python
def max_sum_subarray(nums, k):
    window = [0,k]
    max_sum = 0
    sub_array = nums[window[0]:window[1]]
    sum_sub_array = sum(sub_array)
    max_sum = sum_sub_array
    while window[1] < len(nums):
        sum_sub_array = sum_sub_array - nums[window[0]]
        window[0] += 1
        sum_sub_array = sum_sub_array + nums[window[1]]
        window[1] += 1
        if max_sum < sum_sub_array:
            max_sum = sum_sub_array
    return max_sum
```

```python
def lengthOfLongestSubstring(s: str) -> int:
    window = [0,0]
    char_index_map = {}
    curr_length = 0
    max_length = 0
    for index,char in enumerate(s):
        if char not in char_index_map:
            char_index_map[char] = index
            window[1] += 1
        else:
            window[0] = char_index_map[char] + 1 if char_index_map[char] >= window[0] else window[0]
            window[1] += 1
            char_index_map[char] = index
        curr_length = window[1] - window[0] 
        max_length = max(curr_length,max_length)
    return max_length
```

```python
import heapq


class MinHeap:
    def __init__(self, iterable=None):
        self._data = []
        if iterable is not None:
            self._data = list(iterable)
            heapq.heapify(self._data)

    def push(self, x):
        heapq.heappush(self._data, x)

    def pop(self):
        if not self._data:
            raise IndexError("pop from empty heap")
        return heapq.heappop(self._data)

    def peek(self):
        if not self._data:
            raise IndexError("peek from empty heap")
        return self._data[0]

    def __len__(self):
        return len(self._data)

    def __bool__(self):
        return bool(self._data)

    def __repr__(self):
        return f"MinHeap({self._data})"
```

```python
import math


def print_heap_tree(heap):
    arr = heap._data
    n = len(arr)
    if n == 0:
        print("[empty]")
        return

    # Fixed width for each node (based on largest element)
    node_width = max(len(str(x)) for x in arr)
    height = math.floor(math.log2(n))  # 0-based height of last level

    def fmt(x):
        return str(x).center(node_width)

    for level in range(height + 1):
        # indices for this level
        start = 2**level - 1
        end = min(2**(level + 1) - 1, n)

        # spacing: big gaps at top, smaller lower down
        level_nodes = end - start
        if level_nodes <= 0:
            break

        # base gap shrinks each level
        gap = 2 ** (height - level + 1)
        first_indent = 2 ** (height - level) - 1

        # print node values
        line = " " * first_indent
        parts = []
        for i in range(start, end):
            parts.append(fmt(arr[i]))
        line += (" " * gap).join(parts)
        print(line)

        # last level → no children to draw
        if level == height:
            break

        # print slashes under nodes that have children
        slash_line = ""
        for _ in range(first_indent):
            slash_line += " "

        for i in range(start, end):
            left = 2 * i + 1
            right = 2 * i + 2

            left_char = "/" if left < n else " "
            right_char = "\\" if right < n else " "

            # put children slashes with some gap
            slash_line += left_char
            for _ in range(node_width - 2 if node_width >= 2 else 0):
                slash_line += " "
            slash_line += right_char

            # gap between parent groups
            for _ in range(gap - 1):
                slash_line += " "

        print(slash_line)
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None


def build_tree():
    # level 0
    root = Node(10)

    # level 1
    root.left = Node(5)
    root.right = Node(20)

    # level 2
    root.left.left = Node(3)
    root.left.right = Node(7)
    root.right.left = Node(15)
    root.right.right = Node(30)

    # level 3
    root.left.right.left = Node(6)

    return root


def print_pretty(root):
    lines, *_ = _display_aux(root)
    for line in lines:
        print(line)


def _display_aux(node):
    # No child
    if node.right is None and node.left is None:
        line = f"{node.val}"
        width = len(line)
        height = 1
        middle = width // 2
        return [line], width, height, middle

    # Only left child
    if node.right is None:
        lines, n, p, x = _display_aux(node.left)
        s = f"{node.val}"
        u = len(s)
        first_line = (x + 1) * " " + (n - x - 1) * " " + s
        second_line = x * " " + "/" + (n - x - 1 + u) * " "
        shifted_lines = [line + u * " " for line in lines]
        return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

    # Only right child
    if node.left is None:
        lines, n, p, x = _display_aux(node.right)
        s = f"{node.val}"
        u = len(s)
        first_line = s + x * " " + (n - x) * " "
        second_line = (u) * " " + "\\" + (n - x - 1) * " "
        shifted_lines = [u * " " + line for line in lines]
        return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

    # Two children
    left, n, p, x = _display_aux(node.left)
    right, m, q, y = _display_aux(node.right)
    s = f"{node.val}"
    u = len(s)
    first_line = (x + 1) * " " + (n - x - 1) * \
        " " + s + y * " " + (m - y) * " "
    second_line = x * " " + "/" + \
        (n - x - 1 + u + y) * " " + "\\" + (m - y - 1) * " "
    if p < q:
        left += [" " * n] * (q - p)
    elif q < p:
        right += [" " * m] * (p - q)
    zipped_lines = zip(left, right)
    lines = [a + u * " " + b for a, b in zipped_lines]
    return [first_line, second_line] + lines, n + m + u, max(p, q) + 2, n + u // 2
```

```python
from typing import List, Dict


def contains_duplicate(nums: List[int]) -> bool:
    """Return True if any value appears at least twice in the list."""
    
    dis:Dict[int.int] = {}
    
    for i in nums:
        if i not in dis:
            dis[i] = 1
        else:
            return False
    
    return True
```

```python
def run_commands(commands):
    pl = Playlist()

    for line in commands:
        parts = line.strip().split()
        if not parts:
            continue

        cmd = parts[0]

        if cmd == "ADD_FRONT":
            pl.add_front(parts[1])
        elif cmd == "ADD_END":
            pl.add_end(parts[1])
        elif cmd == "ADD_AFTER":
            # ADD_AFTER y x
            y, x = parts[1], parts[2]
            pl.add_after(y, x)
        elif cmd == "REMOVE":
            pl.remove(parts[1])
        elif cmd == "PLAY_NEXT":
            pl.play_next()
        elif cmd == "PLAY_PREV":
            pl.play_prev()
        elif cmd == "CURRENT":
            curr = pl.get_current()
            print(curr if curr is not None else "NONE")
        elif cmd == "PRINT_ALL":
            pl.print_all()
        else:
            # unknown command
            pass

    return pl
```

```python
def longest_subarray_sum_k(prices: List[int], k: int) -> int:
    # implement here
    total_sum = sum(prices)
    counter = len(prices)
    for item in prices[::-1]:
        
        if total_sum == k:
            break
        else:
            total_sum = total_sum - int(item)
            counter = counter - 1
    return counter
```

```python
import statistics
def findMaxAverage(nums: List[int], k: int) -> float:
    l,r = 0,0+k
    max_mean = float("-inf")
    
    if len(nums) == k:
        max_mean = max(max_mean,statistics.mean(nums))
        return(float(max_mean))
    
    while r <= len(nums):
        max_mean = max(max_mean,statistics.mean(nums[l:r]))
        l += 1
        r += 1
    
    return max_mean
```

```python
def removeDuplicates(s: str, k: int) -> str:
    chars = []
    counts = []

    for ch in s:
        # if same as last character
        if chars and chars[-1] == ch:
            counts[-1] += 1
        else:
            chars.append(ch)
            counts.append(1)

        # if we hit k duplicates, remove them
        if counts[-1] == k:
            chars.pop()
            counts.pop()

    # rebuild final string
    result = ""
    for i in range(len(chars)):
        result += chars[i] * counts[i]

    return result
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


def addTwoNumbers(l1: ListNode, l2: ListNode) -> ListNode:
    Digits_l1 = 0
    Digits_l2 = 0

    # temp heads just for counting
    temp_head_l1 = l1
    temp_head_l2 = l2

    while temp_head_l1 is not None:
        Digits_l1 += 1
        temp_head_l1 = temp_head_l1.next

    while temp_head_l2 is not None:
        Digits_l2 += 1
        temp_head_l2 = temp_head_l2.next

    def add_numbers(longer: ListNode, shorter: ListNode) -> ListNode:
        carry = 0

        dummy = ListNode()   # use as dummy head
        tail = dummy

        l1 = longer
        l2 = shorter

        while l1 is not None:
            v1 = l1.val
            v2 = l2.val if l2 is not None else 0

            current_digit = v1 + v2 + carry
            carry = current_digit // 10
            current_digit = current_digit % 10

            # always create node
            new_node = ListNode(current_digit)
            tail.next = new_node
            tail = tail.next

            l1 = l1.next
            if l2 is not None:
                l2 = l2.next

        if carry > 0:
            tail.next = ListNode(carry)

        return dummy.next

    if Digits_l1 > Digits_l2:
        output_list = add_numbers(l1, l2)
    elif Digits_l2 > Digits_l1:
        output_list = add_numbers(l2, l1)
    else:  # Digits_l1 == Digits_l2
        output_list = add_numbers(l1, l2)

    return output_list
```

```python
# Helper functions to build and print lists
def build_list(arr):
    dummy = ListNode(0)
    curr = dummy
    for num in arr:
        curr.next = ListNode(num)
        curr = curr.next
    return dummy.next


def print_list(head):
    out = []
    while head:
        out.append(head.val)
        head = head.next
    print(out)


# SAMPLE TESTS
l1 = build_list([2, 4, 3])
l2 = build_list([5, 6, 4])

res = addTwoNumbers(l1, l2)
print_list(res)   # expected [7,0,8]


# Additional tests:

# Test 2
l1 = build_list([0])
l2 = build_list([0])
res = addTwoNumbers(l1, l2)
print_list(res)   # expected [0]

# Test 3
l1 = build_list([9, 9, 9, 9, 9, 9, 9])
l2 = build_list([9, 9, 9, 9])
res = addTwoNumbers(l1, l2)
print_list(res)   # expected [8,9,9,9,0,0,0,1]
```

```python
from itertools import chain

def searchMatrix(matrix: List[List[int]], target: int) -> bool:
    new_array = list(chain.from_iterable(matrix))
    
    def binary_search(list:List[int]):
        if not list:
            return False
        mid = len(list) // 2
        # print(f"list = {(list)}, mid = {mid}")
        if target == list[mid]:
            return True
        else:
            if (target > list[mid]):
                # print()
                return binary_search(list[mid+1:])
            else:
                return binary_search(list[0:mid])
    
    return binary_search(new_array)
    
```

```python
import heapq
class KthLargest:
    
    def __init__(self, k: int, nums: List[int]):
        self.nums = nums
        self.k = k
        heapq.heapify(self.nums)
        
        while len(self.nums) > k:
            heapq.heappop(self.nums)
        
        

    def add(self, val: int) -> int:
        heapq.heappush(self.nums,val)
        if len(self.nums) > self.k:
            heapq.heappop(self.nums)
            
        return self.nums[0]
        
```

```python
k = 3
nums = [4, 5, 8, 2]

kth = KthLargest(k, nums)

print(kth.add(3))   # expected 4
print(kth.add(5))   # expected 5
print(kth.add(10))  # expected 5
print(kth.add(9))   # expected 8
print(kth.add(4))   # expected 8

```

```python
def lengthOfLongestSubstring(s: str) -> int:
    left,right = 0,0
    indexPos = {}
    maxLength = 0
    
    if s == "":
        return 0
    
    while right < len(s):
        
        currentCharacter = s[right] 
        currentIndex =  right
        
        if ((currentCharacter in indexPos) and (indexPos[currentCharacter] >= left)): 
                left = indexPos[currentCharacter] + 1
                
        indexPos[currentCharacter] = currentIndex
                
        maxLength = max(maxLength,(right-left+1))
        right += 1

    return maxLength
        
 
```

```python
class Node:
    def __init__(self,key:str = "",value:int=0):
        self.key = key
        self.value = value
        self.next = None
        self.prev = None


class QuoteCache:
    

    def __init__(self, capacity: int):
        self.cache = {}
        self.cache_max_size = capacity
        
        
        self.head = Node()
        self.tail = Node()
        
        self.head.next = self.tail
        self.tail.prev = self.head
        
        
        self.curr = None
        
        # head <-> tail
        
    
        
        
    def check_capacity(self) -> bool:
        return len(self.cache) == self.cache_max_size
    
    def add_to_list(self,key:str,value:int) -> Node:
        new_node:Node = Node(key=key,value=value)
        
        if not self.curr:
            self.head.next = new_node
            self.tail.prev = new_node
            
            new_node.prev = self.head
            new_node.next = self.tail
            
            self.curr = new_node
            # head <-> node1 <-> tail
        else:
            self.curr.next = new_node
            self.tail.prev = new_node
            
            new_node.prev = self.curr
            new_node.next = self.tail
            
            self.curr = new_node
            
            # head <-> node1 <-> node2 <-> tail
            #.           t         c
        return new_node
            
            

    def remove_particular_node(self,toBeRemovedNode:Node) -> Node:
        self.curr = toBeRemovedNode
        
        if self.head.next == self.tail:
            # do nothing since no node there or cache size 0
            return None
        
        if (self.curr.prev == self.head and self.curr.next == self.tail):
            self.head.next = self.tail
            self.tail.prev = self.head
            self.curr = None

            toBeRemovedNode.next = None
            toBeRemovedNode.prev = None

            return toBeRemovedNode
        else:
            self.curr = toBeRemovedNode.next
            self.curr.prev = toBeRemovedNode.prev
            
            toBeRemovedNode.prev.next = self.curr
            
            toBeRemovedNode.next = None
            toBeRemovedNode.prev = None
            
            self.curr = self.tail.prev
            
            return toBeRemovedNode
            
                

    def get(self, ticker: str) -> int:
            """
            Return the latest price for the given ticker.
            If the ticker is not present in the cache, return -1.
            Accessing a ticker (if present) marks it as most recently used.
            """
            if ticker not in self.cache:
                return -1

            # existing node
            node = self.cache[ticker]

            # remove from current position
            self.remove_particular_node(node)

            # reinsert as MRU (at the end)
            new_node = self.add_to_list(key=node.key, value=node.value)

            # update mapping to new node
            self.cache[ticker] = new_node

            return new_node.value

            
        

    def put(self, ticker: str, price: int) -> None:
        """
        Insert or update the latest price for the given ticker.
        If the ticker already exists, update its price and mark it as most recently used.
        If inserting a new ticker causes the number of entries to exceed capacity,
        evict the least recently used ticker.
        """
        

        
        if ticker in self.cache:
            # new node to be added to linked list old node to be removed new added node to be updated in cache
            toBeRemovedNode = self.remove_particular_node(
                toBeRemovedNode=self.cache[ticker])
            del self.cache[ticker]
            # print("To be removed",toBeRemovedNode,toBeRemovedNode.key,toBeRemovedNode.value)
            
            
    
        if(self.check_capacity()):
            #deleting LRU key from cache
            del self.cache[self.head.next.key]
            #removing node from list
            removedNode:Node = self.remove_particular_node(self.head.next)

            

        addedNode = self.add_to_list(key=ticker,value=price)
        self.cache[ticker] = addedNode

            
        # print(self.cache, self.head.next.key, self.head.next.value)
        # print(self.cache, self.tail.prev.key, self.tail.prev.value)
        
```

```python
print("\n=== TEST 6: bigger mixed scenario ===")
cache = QuoteCache(3)

cache.put("AAPL", 100)
cache.put("MSFT", 200)
cache.put("GOOG", 300)      # cache = [AAPL, MSFT, GOOG]; LRU=AAPL, MRU=GOOG

print(cache.get("AAPL"))    # 100; order: [MSFT, GOOG, AAPL]

cache.put("TSLA", 400)      # evict MSFT; cache = [GOOG, AAPL, TSLA]

print(cache.get("MSFT"))    # -1
print(cache.get("GOOG"))    # 300
print(cache.get("AAPL"))    # 100
print(cache.get("TSLA"))    # 400

# LRU is GOOG; evict GOOG; cache = [AAPL, TSLA, NFLX]
cache.put("NFLX", 500)

print(cache.get("GOOG"))    # -1
print(cache.get("AAPL"))    # 100
print(cache.get("TSLA"))    # 400
print(cache.get("NFLX"))    # 500

# update AAPL and TSLA, then add AMZN

cache.put("AAPL", 1100)     # update + move to MRU
cache.put("TSLA", 1400)     # update + move to MRU (AAPL still in cache)

print(cache.get("AAPL"))    # 1100
print(cache.get("TSLA"))    # 1400

# capacity=3, evict LRU (whichever hasn't been touched recently: likely NFLX)
cache.put("AMZN", 600)

print(cache.get("NFLX"))    # should be -1 if NFLX was LRU
print(cache.get("AAPL"))    # 1100
print(cache.get("TSLA"))    # 1400
print(cache.get("AMZN"))    # 600
```
