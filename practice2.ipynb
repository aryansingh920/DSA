```python
from typing import List
from collections import Counter

def longestConsecutive(nums: List[int]) -> int:
    hashSet = set(nums)
    
    # print(hashSet)
    
    longestSeqCount = 0
    
    for num in nums:
        currSequence = 0
        tempNum = num
        
        if tempNum - 1 in hashSet:
            continue
        
        while tempNum in hashSet:
            currSequence += 1
            tempNum += 1
            
        longestSeqCount = max(currSequence, longestSeqCount)
        
    return longestSeqCount

    
```

```python
nums = [100, 4, 200, 1, 3, 2]
# nums = range(1,100) 
# 1 2 3 4.   100.   200
longestConsecutive(nums=nums)
```

```python
from typing import List
from collections import Counter
import heapq



def topKFrequent(words: List[str], k: int) -> List[str]:
    CounterDict = Counter(words)
    heapCount = []
    heapq.heapify(heapCount)
    for word,count in CounterDict.items():
        # print(word,count)
        if len(heapCount) < k:
            heapq.heappush(heapCount, (count,word))
        else:
            if(heapCount[0][0] < count or (count == heapCount[0][0] and 
               (heapCount[0][1]) > (word))):
                heapq.heapreplace(heapCount,(count,word))
    
    heapCount.sort(key=lambda x: (-x[0], x[1]))
    return ([n for i,n in heapCount])
```

```python
words = ["the", "day", "is", "sunny", "the", "the","barn", "the", "is","sunny", "is", "is"]
k = 4

topKFrequent(words,k)
```

```python
from typing import List


def maxArea(height: List[int]) -> int:
    left,right = 0,len(height)-1
    max_Area = 0
    while left < right:
        currArea = (right-left) * min(height[left],height[right])
        max_Area = max(currArea,max_Area)
        
        if height[left] <= height[right]:
            left += 1
        else:
            right -= 1
            
            
    return max_Area

```

```python
maxArea(height=[1, 8,6,2,5,4,8,3,7])
```

```python
def lengthOfLongestSubstring(s: str) -> int:
    
    maxLength = 0
    charMap = {}
    window = {"l" : 0,
              "r" : 0}
    
    while window["r"] < len(s):
        currCharIndex = window["r"]
        currChar = s[currCharIndex]
        # print(currChar, currCharIndex, window, charMap)
        
        if currChar in charMap and charMap[currChar] >= window["l"]:
            window["l"] = charMap[currChar] + 1

        charMap[currChar] = currCharIndex
            
        # print(currChar,currCharIndex,window,charMap)
            
        currStr = s[window["l"]:window["r"]+1]
        # print(currStr)
        
        maxLength = max(len(currStr),maxLength)
        window["r"] += 1
    return maxLength


s = "abcabcbb"
print("ans = ",lengthOfLongestSubstring(s=s))

# print()
# for index, char in enumerate(s):
#     print(index, end=" ")
#     # print(char)
# print()
# for index, char in enumerate(s):
#     # print(index, end=" ")
#     print(char, end=" ")
```
